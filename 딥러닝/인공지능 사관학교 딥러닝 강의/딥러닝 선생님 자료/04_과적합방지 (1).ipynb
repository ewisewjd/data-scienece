```python
# /content/drive/MyDrive/data/sonar.csv
import pandas as pd
sonar = pd.read_csv('/content/drive/MyDrive/data/sonar.csv',header = None)
sonar.shape
```

```python
# 데이터 분리
datasets = sonar.values
X = datasets[:, :60]
y = datasets[:, 60]
```

```python
# 문자열로 된 데이터를 숫자로 변경
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(y)
y = e.transform(y)
y
```

```python
X
```

```python
# 훈련과 테스트셋으로 분리, 데이터 섞어주기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# Dropout을 사용하여 과적합 줄이기
# 모델 설계
model1 = Sequential()

# 입력층
model1.add(Dense(units = 120, input_dim = 60, activation = 'relu'))
# 중간층
model1.add(Dense(units = 240, activation = 'relu'))
model1.add(Dropout(0.8))
model1.add(Dense(units = 120, activation = 'relu'))
model1.add(Dropout(0.8))
model1.add(Dense(units = 60, activation = 'relu'))
model1.add(Dropout(0.8))
# 출력층
model1.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# 실행 방법 설정
model1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
import numpy as np
X_train = X_train.astype(np.float32)
```

```python
X_train
```

```python
# 학습
model1.fit(X_train,y_train,epochs = 200, validation_split = 0.3)
```

```python
X_test = X_test.astype(np.float32)
model1.evaluate(X_train,y_train)
```

```python
model1.evaluate(X_test,y_test)
```
