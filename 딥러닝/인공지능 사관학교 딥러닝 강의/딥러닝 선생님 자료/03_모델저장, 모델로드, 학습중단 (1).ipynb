```python
# 1. 문제정의 : wine데이터를 사용해서 wine의 도수를 예측해보자
```

```python
# 2. 데이터 로드
import pandas as pd
wine = pd.read_csv('/content/drive/MyDrive/wine.csv')
wine.head()
```

```python
# 회귀문제
wine['alcohol']
```

```python
# 3. 데이터 전처리
# 도수(alcohol)을 정답으로
# 나머지를 문제로 만들기
y = wine['alcohol']
X = wine.copy()
del X['alcohol']
```

```python
X.shape, y.shape
```

```python
# 4. 탐색적 데이터 분석
# 머신러닝 - 데이터를 직접 살펴보고 특징을 추출해야 분석을 잘했음
# 딥러닝 - 데이터 특징추출 마저도 모델이 할수있음
# 딥러닝에서는 거의 생략
```

```python
# 5. 모델 선택 및 하이퍼 파라미터 튜닝 - 머신러닝
# 5. 모델 설계(Sequential) 및 학습방법(compile) 설정 - 딥러닝
```

```python
from tensorflow.keras.models import Sequential # 딥러닝 틀(구조) 만들기
from tensorflow.keras.layers import Dense # 딥러닝 구조에 퍼셉트론층 쌓기

model1 = Sequential() # 딥러닝 구조 생성

# 입력층 쌓기
# 퍼셉트론수(units), 입력데이터 크기(input_dim), 활성화 함수(activation)
model1.add(Dense(units = 24 , input_dim = 12 , activation = 'relu'))

# 중간층 쌓기
# 퍼셉트론수, 활성화 함수
model1.add(Dense(units = 48, activation = 'relu'))
model1.add(Dense(units = 96, activation = 'relu'))
model1.add(Dense(units = 48, activation = 'relu'))
model1.add(Dense(units = 24, activation = 'relu'))
model1.add(Dense(units = 12, activation = 'relu'))

# 출력층 쌓기
# 퍼셉트론수(결과 데이터의 수), 활성화 함수
# 회귀 : 1개
# 이진분류 : 1 or 2개
# 다중분류 : class 개수
model1.add(Dense(units = 1, activation = 'linear'))

model1.summary()
```

```python
# 학습방법 설정
# loss : mse
# optimizer : adam
# metrics : mse > 기본값이므로 생략 가능
model1.compile(
    loss = 'mse',
    optimizer = 'adam',
    metrics = ['mse']
)
```

```python
# 6. 학습
# 문제데이터, 정답데이터, batch_size, 학습횟수
model1.fit(X , y , batch_size = 100, epochs = 100 )
```

```python
# 7.평가
model1.evaluate(X,y)

# 총 5개의 층 활성화함수 : sigmoid > loss =1.42
# 총 7개의 층 활성화함수 : sigmoid > loss =1.42
# 총 7개의 층 활성화함수 : relu > loss =0.75
```

```python
# 분류 실습 화이트와인(0)과 레드와인(1) 구분하기
wine['type'].unique()
y = wine['type']
X = wine.copy()
del X['type']
```

```python
X.shape, y.shape
```

```python
from tensorflow.keras.models import Sequential # 딥러닝 틀(구조) 만들기
from tensorflow.keras.layers import Dense # 딥러닝 구조에 퍼셉트론층 쌓기

model2 = Sequential() # 딥러닝 구조 생성

# 입력층 쌓기
# 퍼셉트론수(units), 입력데이터 크기(input_dim), 활성화 함수(activation)
model2.add(Dense(units = 24 , input_dim = 12 , activation = 'relu'))

# 중간층 쌓기
# 퍼셉트론수, 활성화 함수
model2.add(Dense(units = 48, activation = 'relu'))
model2.add(Dense(units = 96, activation = 'relu'))
model2.add(Dense(units = 48, activation = 'relu'))
model2.add(Dense(units = 24, activation = 'relu'))
model2.add(Dense(units = 12, activation = 'relu'))

# 출력층 쌓기
# 퍼셉트론수(결과 데이터의 수), 활성화 함수
# 회귀 : 1개 : activation : linear
# 이진분류 : 1 or 2개 activation : sigmoid / softmax
# 다중분류 : class 개수 activation : softmax
model2.add(Dense(units = 1, activation = 'sigmoid'))

model2.summary()
```

```python
# 학습방법 설정
# loss : binary_crossentropy = 이진분류시 사용하는 오차(출력층 활성화함수가 sigmoid일때만 사용)
# optimizer : adam
# metrics : accuracy
model2.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
# 6. 학습
# 문제데이터, 정답데이터, batch_size, 학습횟수
model2.fit(X , y , batch_size = 100, epochs = 100 )
```

```python
# 7. 평가
model2.evaluate(X, y)
```

```python
# 모델 저장
# 가장 마지막 학습의 모델을 저장
model2.save('/content/drive/MyDrive/model/wine_model.h5')
```

```python
from tensorflow.keras.models import Sequential # 딥러닝 틀(구조) 만들기
from tensorflow.keras.layers import Dense # 딥러닝 구조에 퍼셉트론층 쌓기

model2 = Sequential() # 딥러닝 구조 생성

# 입력층 쌓기
# 퍼셉트론수(units), 입력데이터 크기(input_dim), 활성화 함수(activation)
model2.add(Dense(units = 24 , input_dim = 12 , activation = 'relu'))

# 중간층 쌓기
# 퍼셉트론수, 활성화 함수
model2.add(Dense(units = 48, activation = 'relu'))
model2.add(Dense(units = 96, activation = 'relu'))
model2.add(Dense(units = 48, activation = 'relu'))
model2.add(Dense(units = 24, activation = 'relu'))
model2.add(Dense(units = 12, activation = 'relu'))

# 출력층 쌓기
# 퍼셉트론수(결과 데이터의 수), 활성화 함수
# 회귀 : 1개 : activation : linear
# 이진분류 : 1 or 2개 activation : sigmoid / softmax
# 다중분류 : class 개수 activation : softmax
model2.add(Dense(units = 1, activation = 'sigmoid'))

model2.summary()
```

```python
# 학습방법 설정
# loss : binary_crossentropy = 이진분류시 사용하는 오차(출력층 활성화함수가 sigmoid일때만 사용)
# optimizer : adam
# metrics : accuracy
model2.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
# 학습 중단 : 모델이 더이상 개선되지않으면 학습을 중단
# 베스트 모델 저장 : 전 epoch보다 좋아지면 저장
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# 베스트 모델을 저장하는 폴더
model_dir = '/content/drive/MyDrive/model/'

# 저장할 파일명
model_file = model_dir + "wine_{epoch:02d}-{accuracy:.4f}.hdf5"

mcp = ModelCheckpoint(
    filepath = model_file, # 파일이 저장될 경로
    monitor = 'accuracy', # 정확도를 기준으로 best모델을 판단
    save_best_only = True, # 모델이 개선되었을때만 저장
    verbose = 1 # 진행 내용 출력
)

es = EarlyStopping(
    monitor = 'accuracy', # 정확도를 기준으로 모델을 판단
    patience = 5 # 5번의 학습동안 모델이 개선되지 않으면 학습을 중단
)
```

```python
# 6. 학습
# 문제데이터, 정답데이터, batch_size, 학습횟수
model2.fit(X , y , batch_size = 100, epochs = 1000 , callbacks = [mcp, es])
```
