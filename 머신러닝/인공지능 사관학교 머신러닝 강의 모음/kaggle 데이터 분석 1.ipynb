### 데이터 분석과 모델링은 어떻게 해야할까 
- 타이타닉 데이터 분석한것을 토대로 절차를 메뉴얼 해보자

1. 데이터 수집
    - 데이터를 수집하는 단계 , 일단 데이터를 불러온다
2. 탐색적 데이터 분석
    - 데이터를 이해하는 단계
    - 결측치,이상치, 오류가 있는지 확인
        - info(), describe(), isnull()과 sum()함수 등을 이용
    - 기술 통계
    - 상관관계
    - 시각화
        - 탐색을 할때는 왠만하면 훈련용(train)데이터를 이용하여 분석을 진행한다.
        - test 데이터를 활용하지 않고 train데이터를 이용해서 시각화/ 탐색 하는게 좋다.
3. 특성공학
    - feature에서 유의미한 컬럼을 추출하거나 병합하는 행위 
    - 비정형데이터(텍스트 데이터 등)는 정형화 시켜주어야한다. 
    
4. 모델링

        - label encoding
        - one-hot encoding
