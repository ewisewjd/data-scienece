# 차원축소 


- 차원 축소: 복잡도를 줄여줄 수 있다.
    - PCA: 분산이 중요! , 분산이 최대가 되는 선을 중심으로 차원축소가 진행된다. 
        - 분산: 전체 데이터의 평균으로부터의 거리 /평균과 개별 관측치의 거리
        - 공분산: 각 특성의 분산을 곱셈한것 , 첫번째와 두번째의 거리를 같이 결합하는것
        - 특성들을 동시에 고려했을때 분산이 최대가 되는 축을 찾아서 차원을 축소시켜준다.
        - 축소를 하더라도 특성들의 정보량을 잘 반영했기 때문에 괜찮을 것이다.
